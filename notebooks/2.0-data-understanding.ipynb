```python
mean = house_data.price.mean()
skewness = house_data.price.skew()
median = house_data.price.median()
std = house_data.price.std()
iqr = house_data.price.describe()['75%'] - house_data.price.describe()['25%']
lower_limit = house_data.price.describe()['75%'] - 1.5 * iqr
upper_limit = house_data.price.describe()['25%'] + 1.5 * iqr

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 8))

sns.histplot(data=house_data, x='price', kde=True, ax=axes[0], bins=30, color='teal')
axes[0].set_title('House Price Hist Distribution')
axes[0].set_xlabel('Price (millions)')
axes[0].set_ylabel('Count')
axes[0].set_facecolor('#eaeaf2')
axes[0].axvline(mean, label='Mean', color='purple', linestyle='dashed', linewidth=2)
axes[0].annotate(f'Skewness = {skewness:.2f}', xy=(0.8, 0.9), xycoords='axes fraction', fontsize=9)
axes[0].annotate(f'Mean = {mean / 1000000 :.2f}', xy=(0.8, 0.82), xycoords='axes fraction', fontsize=9)
axes[0].annotate(f'Std = {std / 1000000 :.2f}', xy=(0.8, 0.74), xycoords='axes fraction', fontsize=9)
axes[0].legend(loc='upper right', title='Legend')
axes[0].grid(True)

sns.boxplot(house_data, y='price', ax=axes[1], color='teal')
axes[1].set_title('House Price Box Distribution')
axes[1].set_ylabel('Price (millions)')
axes[1].set_facecolor('#eaeaf2')
axes[1].annotate(f'Median = {median / 1000000 :.2f}', xy=(0.8, 0.9), xycoords='axes fraction', fontsize=9)
axes[1].grid(True)

plt.tight_layout()

plt.show()
```

##### Multivariate Analysis

![correlaÃ§Ã£o de pearson](..\notebooks\pearson-corr.png)

```python
selected_columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']

labels = ['Price', 'Bedrooms', 'Bathrooms', 'Sqft living', 'Sqft lot', 'Floors', 'Sqft above', 'Sqft basement']

selected_data = house_data[selected_columns]

correlation_matrix = selected_data.corr()

plt.figure(figsize=(12, 5), facecolor='#eaeaf2')

ax = sns.heatmap(data=correlation_matrix,
                xticklabels=labels,
                yticklabels=labels,
                annot=True, 
                cmap='coolwarm',
                center=0,
                vmin=-1,
                vmax=1,
                square=True,
                fmt='.2f',
                linewidths=0.7,
                cbar=True
)   
                
plt.title('Correlation of Housing Features')

plt.tight_layout()

plt.show()
```

```python
selected_columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot']

plt.figure(figsize=(12, 6))

pairplot = sns.pairplot(data=house_data, vars=selected_columns, markers='o', palette=['teal', 'purple'], hue='waterfront')

pairplot.fig.suptitle('Pairwise Comparison of Housing Features by Waterfront Status', y=1.02) 

for ax in pairplot.axes.flatten():
    ax.set_facecolor('#eaeaf2')
    ax.grid(True, linestyle='--', linewidth=0.7)

labels = ['Price (millions)', 'Bedrooms', 'Bathrooms', 'Sqft Living (feetsÂ²)', 'Sqft Lot (feetsÂ²)']

for i, label in enumerate(labels):
    for j, sublabel in enumerate(labels):
        if i == j:
            pairplot.axes[i, j].set_xlabel(label)
            pairplot.axes[i, j].set_ylabel(label)
        else:
            pairplot.axes[i, j].set_xlabel(sublabel)
            pairplot.axes[i, j].set_ylabel(label)

pairplot._legend.set_bbox_to_anchor((0.96, 0.99))
pairplot._legend.set_title('Waterfront Status') 
pairplot._legend.set_frame_on(True)  

plt.tight_layout()

plt.show()
```
